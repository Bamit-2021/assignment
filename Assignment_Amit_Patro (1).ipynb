##Preprocessing

```python
# Impute the missing values in the features by filling with this
missing_train_value=train_df.isnull().sum().sort_values(ascending=False)*100/len(train_df)
```

```python
missing_train_value
```

```python
# Impute the missing values in the features by filling with this
missing_test_value=test_df.isnull().sum().sort_values(ascending=False)*100/len(test_df)
```

```python
missing_test_value
```

```python
missing_label_value=label_df.isnull().sum().sort_values(ascending=False)*100/len(label_df)
```

```python
missing_label_value
```

```python
numerics_1 = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_columns_1=train_df.select_dtypes(include=numerics_1).columns.to_list()
len(numeric_columns_1)
```

```python
object_columns_1=train_df.select_dtypes(include=object).columns
len(object_columns_1)
```

```python
# Converting all the features to objct types
train_df[object_columns_1]=train_df[object_columns_1].astype(str)
```

```python
train_df[object_columns_1].info()
```

```python
# from the train_df dataset table, seperate the boolean values by using this.
boolean_columns_1=[]
alphanumeric_columns_1=[]
for col in object_columns_1:
    if train_df[col][0].isupper()==True:
        boolean_columns_1.append(col)
    else:
        alphanumeric_columns_1.append(col)
```

```python
# converting all to categorical values
train_df[boolean_columns_1]=train_df[boolean_columns_1].astype('category')
```

```python
train_df.info()
```

```python
# In all features imputing the missing values of numerical data by using mean and categorical data by most frequent.
num_imputer=SimpleImputer(strategy='mean')
obj_imputer=SimpleImputer(strategy='most_frequent')
```

```python
numerics_2 = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_columns_2=test_df.select_dtypes(include=numerics_2).columns.to_list()
len(numeric_columns_2)
```

```python
object_columns_2=test_df.select_dtypes(include=object).columns
len(object_columns_2)
```

```python
test_df[object_columns_2]=test_df[object_columns_2].astype(str)
```

```python
test_df[object_columns_2].info()
```

```python
# from the test_df dataset table, seperate the boolean values by using this.
boolean_columns_2=[]
alphanumeric_columns_2=[]
for col in object_columns_2:
    if test_df[col][0].isupper()==True:
        boolean_columns_2.append(col)
    else:
        alphanumeric_columns_2.append(col)
```

```python
# converting all to categorical values
test_df[boolean_columns_2]=test_df[boolean_columns_2].astype('category')
```

```python
test_df.info()
```

```python
num_imputer.fit(train_df[numeric_columns_1])
```

```python
train_df[numeric_columns_1]=num_imputer.transform(train_df[numeric_columns_1])
```

```python
num_imputer.fit(test_df[numeric_columns_2])
```

```python
train_df[numeric_columns_1].isna().sum().sort_values(ascending=False)
```

```python
test_df[numeric_columns_2].isna().sum().sort_values(ascending=False)
```

```python
train_df[boolean_columns_1]
```

```python
# Replace the boolean values with 1,o,None in train_df.
for col in boolean_columns_1:
    train_df[col].replace({'YES':1, 'NO':0,'nan':None},inplace=True)
```

```python
# Replace the boolean values with 1,o,None in test_df.
for i in train_df:
  train_df[i].fillna(method='ffill',inplace=True)
```

```python
train_df[boolean_columns_1]
```

```python
obj_imputer.fit(train_df[boolean_columns_1])
train_df[boolean_columns_1]=obj_imputer.transform(train_df[boolean_columns_1])
```

```python
for col in boolean_columns_2:
    test_df[col].replace({'YES':1, 'NO':0,'nan':None},inplace=True)
```

```python
# Replacing the null values with fillna.
for i in test_df:
  test_df[i].fillna(method='ffill',inplace=True)
```

```python
test_df[boolean_columns_2]
```

```python
obj_imputer.fit(test_df[boolean_columns_2])
test_df[boolean_columns_2]=obj_imputer.transform(test_df[boolean_columns_2])
```

```python
# Converting the datatypes to integer.
train_df[boolean_columns_1]=train_df[boolean_columns_1].astype('int')
test_df[boolean_columns_2]=test_df[boolean_columns_2].astype('int')
```

## Model Building

```python
# Split the datasets with train and test.
xtrain=train_df[numeric_columns_1+boolean_columns_1]
xtest=test_df[numeric_columns_2+boolean_columns_2]
ytrain=label_df.loc[:9998,'y1':]
```

```python
# using the naive bayes model for model creation.
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = BinaryRelevance(classifier=GaussianNB())
```

```python
classifier.fit(xtrain.values,ytrain.values)
```

```python
predictions = classifier.predict(xtest.values)
```
